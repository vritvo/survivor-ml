# Survivor Data — Exploratory Data Analysis

Explore each table loaded in `src/load.py` to understand structure, nulls, and distributions.
Findings here drive the cleaning functions.

```python
import pandas as pd

pd.set_option("display.max_columns", None)
pd.set_option("display.width", 140)

xls = pd.ExcelFile("../data/survivoR.xlsx")

SHEETS = [
    "Castaways", "Castaway Details", "Episodes", "Vote History",
    "Challenge Results", "Confessionals", "Boot Mapping",
    "Advantage Details", "Advantage Movement", "Tribe Mapping",
]
data = {s: pd.read_excel(xls, sheet_name=s) for s in SHEETS}

def explore(name):
    """Print shape, column types, and null counts for a table."""
    df = data[name]
    print(f"Shape: {df.shape[0]:,} rows x {df.shape[1]} cols\n")
    nulls = df.isnull().sum()
    print("Columns:")
    for col in df.columns:
        n = nulls[col]
        pct = f"  — {n:,} nulls ({n/len(df)*100:.0f}%)" if n > 0 else ""
        print(f"  {col:30s} {str(df[col].dtype):12s}{pct}")
    print()
    return df.head(3)
```

## Overview — Table sizes

```python
for name in SHEETS:
    df = data[name]
    print(f"{name:25s} {df.shape[0]:>6,} rows x {df.shape[1]:>2} cols")
```

## Castaways

One row per castaway per season appearance (returnees appear multiple times).
Key table for player metadata per season: age, result, tribe, place.

```python
explore("Castaways")
```

```python
df = data["Castaways"]
print("version (franchise) distribution:")
print(df["version"].value_counts().to_string())
print(f"\nresult values:")
print(df["result"].value_counts().to_string())
print(f"\njury_status values:")
print(df["jury_status"].value_counts(dropna=False).head(10).to_string())
```

## Castaway Details

One row per unique castaway (across all appearances).
Demographics, personality type, occupation.

```python
explore("Castaway Details")
```

```python
df = data["Castaway Details"]
print("gender:")
print(df["gender"].value_counts().to_string())
print(f"\npersonality_type (top 5):")
print(df["personality_type"].value_counts().head().to_string())
print(f"\ncollar:")
print(df["collar"].value_counts(dropna=False).to_string())
print(f"\ndate_of_birth range: {df['date_of_birth'].min()} to {df['date_of_birth'].max()}")
```
